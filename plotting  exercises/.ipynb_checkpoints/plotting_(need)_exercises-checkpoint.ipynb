# Exercise 1

```python
from matplotlib import pyplot as plt
```

```python

x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=.3, yerr=[0,0.5,1.0,1.5,2,2.5],
             color="#8d0505", linestyle="--",
             marker="o", markerfacecolor="#ffff00",
             ecolor="#000000",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = "lower right")
plt.show()
```

# Exercise 2

### Task1

```python
import numpy as np
```

```python
x = np.linspace(0,50)
x10 = -x/10
ex = np.e**x10
sinx = np.sin(x)
y = sinx * ex
plt.plot(x, y, label='y',color='red', linestyle='dashed')
plt.legend()                                 
```

### Task 2

```python
x = np.linspace(0,50,400)
x10 = -x/10
ex = np.e**x10
sinx = np.sin(x)
y = sinx * ex
plt.plot(x, y, label='y',color='blue', linestyle='-')
plt.xlabel('x values', fontsize = 15, color = "darkorange")
plt.ylabel('y values', fontsize = 15, color = "darkorange")
plt.grid(alpha = 0.25)
plt.title(r'Graph of $y = e^{(-x/10)}sin(x)$', fontsize=20)
plt.legend()    
plt.ylim([-1,1])
plt.show()
```

### Alternative solution:

```python
x = np.linspace(0,50,400)
y = np.exp(-x/10) * np.sin(x)
plt.plot(x, y, label='y',color='blue', linestyle='-')
plt.xlabel('x values', fontsize = 15, color = "darkorange")
plt.ylabel('y values', fontsize = 15, color = "darkorange")
plt.grid(alpha = 0.25)
plt.title(r'Graph of $y = e^{(-x/10)}sin(x)$', fontsize=20)
plt.legend()    
plt.ylim([-1,1])
plt.show()
```
