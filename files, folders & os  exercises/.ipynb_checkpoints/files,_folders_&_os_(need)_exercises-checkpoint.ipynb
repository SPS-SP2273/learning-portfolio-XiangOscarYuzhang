## The scenario

Before doing the task, since the folder have a lot of unwanted files, we can delete them so that we don't need to filter later. However, deleting the original file is very dangerous; thus, first copy a backup and then delete.

```python
import shutil
import os
```

```python
os.makedirs("data_back_up")  #create a new folder to back up the data since there will be some changing later.
```

```python
source_folder = r'C:\Users\oscar\Documents\GitHub\learning-portfolio-XiangOscarYuzhang\files, folders & os  exercises\os-collaboration-exercise-data'
back_up_folder = r'C:\Users\oscar\Documents\GitHub\learning-portfolio-XiangOscarYuzhang\files, folders & os  exercises\data_back_up'
shutil.copytree(source_folder,back_up_folder)
#copy the folder
```

```python
deleting_data_path = r'C:\Users\oscar\Documents\GitHub\learning-portfolio-XiangOscarYuzhang\files, folders & os  exercises\data_back_up'

for file in os.listdir(deleting_data_path):  #for loop created to do repeating task
    if file.endswith('.png'): #checking if the file is ending with png
        file_path = os.path.join(deleting_data_path,file) #locate the actual path of that png file
        print(file_path) # testing before remove
        os.remove(file_path) #deleting all the png file
    if file.endswith('.jpg'): #checking if the file is ending with jpg
        file_path = os.path.join(deleting_data_path,file) #locate the actual path of that jpg file
        print(file_path) # testing before remove
        os.remove(file_path) #deleting all the jpg file
    if file.endswith('.pdf'): #checking if the file is ending with pdf
        file_path = os.path.join(deleting_data_path,file) #locate the actual path of that pdf file
        print(file_path) # testing before remove
        os.remove(file_path)
```
